```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
import matplotlib.patches as patches

def create_split_animation():
    # Set up the figure
    fig, axs = plt.subplots(3, 4, figsize=(15, 12))
    axs = axs.ravel()

    # Data for different splitting patterns with multiple columns where needed
    patterns = {
        'holdout': {'data': [[1]*7 + [2]*3], 'title': '1. holdout'},
        'inv_holdout': {'data': [[2] + [1]*9], 'title': '2. inv_holdout'},
        'rep_holdout': {'data': [[1,1,1,2,2,2,1,1,1,2], [1,1,1,2,2,2,1,1,1,2],
                               [1,1,1,2,2,2,1,1,1,2], [1,1,1,2,2,2,1,1,1,2]],
                       'title': '3. rep_holdout'},
        'cv': {'data': [[1,2,1,1,2,1,2,1,1,2], [1,1,2,1,2,1,1,2,1,2],
                       [1,1,1,2,1,2,1,1,2,1], [1,1,1,1,2,1,2,1,2,2]],
               'title': '4. cv'},
        'cv_mod': {'data': [[1,2,0,2,1,0,2,1,0,2], [1,1,2,0,2,1,0,2,1,0],
                          [0,1,1,2,0,2,1,0,2,1], [1,0,1,1,2,0,2,1,0,2]],
                  'title': '5. cv_mod'},
        'cv_bl': {'data': [[2,1,1,1,2,1,1,2,1,2], [1,2,1,1,1,2,1,1,2,1],
                         [1,1,2,1,1,1,2,1,1,2], [1,1,1,2,1,1,1,2,1,2]],
                 'title': '6. cv_bl'},
        'cv_hvbl': {'data': [[2,1,1,1,2,1,2,1,0,2], [1,2,1,1,1,2,1,2,1,0],
                           [1,1,2,1,1,1,2,1,2,1], [1,1,1,2,1,1,1,2,1,2]],
                   'title': '7. cv_hvbl'},
        'preq_bls': {'data': [[0,0,0,2,1,0,0,2,1,2], [0,0,0,0,2,1,0,2,1,2],
                            [0,0,0,0,0,2,1,0,2,1]],
                    'title': '8. preq_bls'},
        'preq_sld_bls': {'data': [[1,0,0,0,2,0,0,2,1,2], [0,1,0,0,0,2,0,2,1,2],
                                [0,0,1,0,0,0,2,0,2,1]],
                        'title': '9. preq_sld_bls'},
        'preq_bls_gap': {'data': [[1,0,0,2,1,2], [0,1,0,0,2,1]],
                        'title': '10. preq_bls_gap'},
        'preq_slide': {'data': [[1,1,1,1,2,2,2,2], [1,1,1,1,2,2,2,2],
                              [1,1,1,1,2,2,2,2], [1,1,1,1,2,2,2,2]],
                      'title': '11. preq_slide'},
        'preq_grow': {'data': [[2,2,2,2,1,1,1,1], [2,2,2,2,1,1,1,1],
                             [2,2,2,2,1,1,1,1], [2,2,2,2,1,1,1,1]],
                     'title': '12. preq_grow'}
    }

    colors = {0: 'white', 1: 'lightblue', 2: 'salmon'}
    max_blocks = max(max(len(col) for col in pattern['data']) for pattern in patterns.values())

    def init():
        for ax in axs:
            ax.clear()
            ax.set_xlim(-0.1, 1.1)
            ax.set_ylim(-0.1, 10.1)
            ax.set_aspect('equal')
            ax.axis('off')

    def draw_frame(frame_num):
        for idx, (pattern_name, pattern) in enumerate(patterns.items()):
            ax = axs[idx]
            if frame_num == 0:
                ax.clear()
                ax.set_title(pattern['title'])
                ax.set_xlim(-0.5, len(pattern['data'])-0.5)
                ax.set_ylim(-0.1, 10.1)
                ax.set_aspect('equal')
                ax.axis('off')

            # Draw blocks up to current frame
            for col_idx, column in enumerate(pattern['data']):
                for row_idx, val in enumerate(column):
                    if row_idx <= frame_num:
                        color = colors[val]
                        rect = patches.Rectangle((col_idx, 9-row_idx), 1, 1,
                                              facecolor=color, edgecolor='black')
                        ax.add_patch(rect)

    anim = FuncAnimation(fig, draw_frame, frames=10,
                        init_func=init, interval=500, repeat=True)

    plt.tight_layout()

    # Save animation
    writer = PillowWriter(fps=2)
    anim.save('cv_patterns_simultaneous.gif', writer=writer)
    plt.close()

# Create the animation
create_split_animation()
```
